```python
km = KMeans(n_clusters=3, random_state=42)
y_predicted = km.fit_predict(df[['petal length (cm)', 'petal width (cm)']])
```

```python
# Add predictions to DataFrame
final_df = df.copy()
final_df['cluster'] = y_predicted
```

```python
final_df1 = final_df[final_df.cluster == 0]
final_df2 = final_df[final_df.cluster == 1]
final_df3 = final_df[final_df.cluster == 2]

# Plot clusters
plt.scatter(final_df1['petal length (cm)'], final_df1['petal width (cm)'], color='green', label='Cluster 0')
plt.scatter(final_df2['petal length (cm)'], final_df2['petal width (cm)'], color='red', label='Cluster 1')
plt.scatter(final_df3['petal length (cm)'], final_df3['petal width (cm)'], color='blue', label='Cluster 2')

# Plot centroids
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='black', marker='+', s=200, label='Centroids')

# Labels and legend
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.title('KMeans Clustering on Petal Length vs Width')
plt.show()
```
